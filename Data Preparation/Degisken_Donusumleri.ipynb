# Değişken Dönüşümleri

```python
# değişken dönüşümleri > standardizasyon. yani her bir standardizasyon, zaten değişken dönüşümüdür. Kafa karışıklığı olmamalı. şöyle düşünülmeli:
# hepsi bir dönüşüm işlemi olarak düşünülmeli. bu işim önemli noktası, değişkenleri dönüştürürken bu değişkenlerin mevcut taşıdığı bilginin yapısını
# bozuyor muyuz? bozmuyor muyuz? bunu göz önünde bulundurmamız lazım. bunu göz önünde bulundurduktan sonra istediğiniz gibi isimlendirebilirsiniz.
# standartlaştırma işleminin verisetinin genel yapısını, özütündeki yapıyı koruyup da belirli değerlerle onları dönüştürme işlemi olduğunu bilmeliyiz.
```

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

### 0-1 Dönüşümü

```python
# bu nedir? örneğin: buradaki cinsiyet değişkenini, 0-1 değerlerine dönüştürmek demek. bu dönüşüm label encoder isimli bir fonk ile yapılıyor ve
# bu alandaki kullanılan en sık dönüşümlerden biri. örneğin bir bağımlı değişkenimiz yani hedeflediğimiz bir değişkenimiz olduğunu düşünelim ve buradaki
# gibi cinsiyet değişkeni olduğunu ve iki sınıflı olduğunu düşünelim. bu durumda ve karşılaşılaşabilecek diğer bir çok durumda, fonksiyonların bizden
# beklemiş olduğu bir format bu. yani diyoruz ki: elimdeki değişken kategorik değişken olabilir, çeşitli sınıf değerlerinden oluşuyo olabilir, ama
# ben bir fonk olarak bunu böyle istemiyorum, bana bunu nümerik bir formatta gönder demek.
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
lbe.fit_transform(df["sex"])
```

```python
df #orginal halinde bu şekilde olmadı. genelde, bu yeni yapılan dönüştüme işlemini de verisetine eklemek gerekiyor. bunu şöyle yapabiliriz:
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"]) #eskisi de kalsın yenisini de ekleyelim durumu.
```

```python
df #kritik uyarı: örneğin 1-0 dönüşümünü yaptığımızda, hangi kategorik değişkenin değerine hangi nümerik değerin geldiğini iyi anlamak gerekiyor.
# çünkü ileride görecek olduğumuz bir sınıflandırma problemi ele aldığımızda, mesela titanic veristeinde insanların hayatta kalıp kalamaması, burada
# genelde, ilgilendiğimiz ana odağımızı 1 sınıfı olarak işaretliyoruz. yani hayatta kalmayı 1 olarak işaretliyoruz.
```

### "1 ve Diğerleri (0)" Dönüşümü

```python
# örneğin verisetimiz içerisinden ikiden fazla sınıf var. bunların arasından seçmiş olacağımız bir tanesine 1,
# diğerlerine 0 sıfır dediğimiz durum. örneğin 3-4 sınıflı hedef değişken, bağımlı değişken var. bunun 1 tanesi ya da 
# diğerleri şeklinde bir odaklanma ihtiyacı olduğunda, bu dönüşümü kullanabiliyoruz.
```

```python
df.head() #mesela burada day veya time değişkeni için böyle bir dönüştürme işlemi gerçekleştirilebilir.
```

```python
import numpy as np
df["yeni_day"] = np.where(df["day"].str.contains("Sun"), 1, 0) # yani dedik ki: birkoşul çalıştırmak istiyoruz.
# day içerisindeki değerleri string olarak değerlendir ve sunday içerenlere 1, diğerlerine 0 demek istiyoruz dedik. 
```

```python
df #nasıl anlayamadık mı? şöyle yapıcaz anlamadığımızda mesela şu kısmı anlamadık: df["day"].str.contains("Sun")
```

```python
df["day"].str.contains("Sun") #belirli koşulu sağlayığ sağlanmadığını sorgulayan bu tip çıktılar ile uğraştık.
# where şunu sağlıyor: df["day"].str.contains("Sun") bu koşulu sağlayanlara 1 ver, gerisine 0 ver. sun içeriyorsa demek. 
```

```python
#sınıf sayısı kadar 0'dan başlacak şekilde, bir dönüştürme işlemi gerçekleştirdik.
# ÇOK DİKKAT: eğer elimizdeki kategorik değişkenin 2'den fazla sınıfı varsa, ve fonksiyonların bizden beklentisi
# doğrultusunda bu kategorik değişkeni bu şekilde nümerik bir formata çevirmek istiyorsak, neticesinde böyle bir çevirme
# işlemi yaptıysak; (burası çok hassas bir nokta) eğer elimizdeki bir değişkeni bu şekilde bir nümerik değerlere
# dönüştür de, algoritmalara gönderirsek, bu artık bunu 0 ile 3 arasındaki değerlerden oluşan bir değişken olarak
# tanımlayacak. yani, bu dönüşüm algoristmaların kafasını karıştıracak. ve kategorik değişkenin sınıflarının, bağımlı
# değişkene olan etkileri bozulacaktır. neden? çünkü, burada sınıflar nümerik değerlere dönüştü. normlade bir kategorik
# değişkenin nominal ölçek türüyle ölçülmüş olan durumu, yani sınıflar arasındaki mesafesi eşit olan durumu, bir anda
# bir dönüşüm işlemi ile bozulup, sanki aralarında fark varmışçasına bir sıralama işlemine tabi tutuldu.
# mesela burada pazartesiye 0, salıya 1 dedi. salı pazartesiden büyük mü? hayır. o halde bu durum yanlılığa. gürültüye 
# vs. sebep olabilmekte. bu durumda ne yapmalı? one hot dönüşümü yapmalı.
```

### One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
# nedir? kategorik değişkenleri, nümerik değişkenlere dönüştürdüğümüzde ortaya çıkan bozukluklardan bahsettik.
# bu problemi gidermek için one-hot dönüşümü yapmalıyız. bunu yaparken de dummy tuzağı ortaya çıkıyor bunu nasıl
# atlatırız ondan da bahsedeceğiz. 
```

```python
df.head()
```

```python
import pandas as pd
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"])
# prefix şu demek: oluşturulacak olan değişkenlerin ön isimlendirilmesinin ne olacağını belirtiyoruz. df in içinden cinsiyet değişkenini dönüştür,
# bunu onehot df olarak kaydet dedik. peki ne oldu? elimizdeki iki sınıflı kategorik değişkenin sınıf sayısı kadar yeni değişken oluştu. ve oluşan bu 
# yeni değişkenler de 1-0 olarak tanımlanmış oldu. mesela ilk gözlem biriminde kadın 1.ikinci gözlem birimi erkekç o halde 0 olacak. bu dönüşüme 
# one hot encoding dönüşümü deniyor. bu dönüşüm elimizdeki kategorik değişkenleri, eğer nümerik değişkenlere dönüştürmek istiyorsak, bu şekilde
# dönüştürmekte. burada şöyle bir problem söz konusu: dummy değişken tuzağı. nedir bu? eğer veriseti içerisindeki değişkenlere bir dönüşüm 
# uyguladığımızda eğer bu dönüşüm sonrasında oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa bu durum dummy değişken tuzağı
# olarak adlandırılır. diğer bir deyişle, bir değişkeni ifade eden başka bir değişken varsa bu duruma dummy değişken tuzağı denir. mesela burada:
# cinsiyet_erkek üzerinden cinsiyet_kadın değişkeni oluşturulabilmektedir. (yani cinsiyet_erkek değişkenindeki sıfırlar zaten kadın demek). yani aslında
# bu iki değişken de aynı bilgiyi taşımaktadır. hatta bazı fonksiyonlar, dönüştürülen kategorik değişkenin original halini de verisetinde tutmakta. 
# bu durumda hem original hali, hem one hot dönüştürülmüş hali (ki iki değişken bu mesela), 3 ü de aynı şeyi ifade ediyor oluyor. işte bu duruma
# dummy değişken tuzağı deniyor. bu durumu nasıl kontrol altında tutabiliriz? şöyle: oluşturulan değişken sayısından bir az sayıda, daha doğru ifadesi
# kategorik değişkenin sınıf sayısından daha az sayıda, dummy değişken olması gerekiyor. örneğin burada cinsiyetin iki tane sınıfı vardı. bu durumda
# bir tane dummy değişken olmalı. değişkenin original hali de verisetinde olmamalı. çünkü iki değişken de aynı bilgiyi ifade ediyor.
```
