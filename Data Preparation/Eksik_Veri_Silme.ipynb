# Eksik Veri Silme Yöntemleri

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}
) #3 adet numpy array oluşturup, bunlardan bir dataframe oluşturduk.
df
```

```python
df.dropna() #en az 1 tane eksik değer olanı sildik.
```

```python
df.dropna(how = "all") #hepsi eksik ise o gözlemi sil (6 gitti)
```

```python
#peki, gözlemler bazında silme işlemini yaptık. Değişkenler bazında silme işlemi yapmak istersek?
```

```python
df.dropna(axis = 1) # dedik ki sütun olarak, en az 1 eksik varsa sil, hepsi gitti zira her değişken için en az bir eksik var zaten
```

```python
df.dropna(axis = 1, how = "all") # tüm değerleri Na olan değişkeni silecek. Tabi hiç silme olmadı zira Hepsi Na olan değişken yok
```

```python
df["Silinecek"] = np.nan # her gözlemde boş olan bir değişken oluşturduk
```

```python
df
```

```python
df.dropna(axis = 1, how = "all") # şimdi silmiş oldu
```

```python
df.dropna(axis = 1, how = "all", inplace = True) #dersek, kalıcı olarak kayıt eder.
```
