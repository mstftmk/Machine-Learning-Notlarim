# Eksik Veri Değer Atama

### Sayısal Değişkenlerde Atama

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}
) #3 adet numpy array oluşturup, bunlardan bir dataframe oluşturduk.
df
```

```python
df["V1"].fillna(0) #eksikleri 0 ile doldur
```

```python
df["V1"].fillna(df["V1"].mean()) #eksikleri ortalama ile doldurduk
```

```python
#Tüm değişkenler için birinci yok
df.apply(lambda x: x.fillna(x.mean()), axis = 0)#Tüm değişkenleri kendi ortalamasıyla doldurduk
```

```python
#Tüm değişkenler için ikinci yol
df.fillna(df.mean()[:]) # fillna i df de çalıştırınca, tüm değişkenlerde bi doldurma işleme yapacağını biliyor. çünkü vektörel fonk. 
#neye göre dolduracak? df içinde mean çalıştırıyoruz her bir değişkenin ort hesaplatırız [:] tüm değişkenleri göz önünde bulundur demek
```

```python
#mesela, dağılımlarına baktık değişkenlerinin, gördük ki v1 ve v2 değişkeni dağılımları düzgün. simetrik dağılımları, normal dağılım eğiliminde.
# yani bunları ortalamayla doldurabiliriz. ama dağılımları düzgün değilse, yapısal bir bozukluk varsa, bunları ortalama ile 
# doldurmak zaten mantıksız olur. bu durumda bu fillna mean işe yaramaz. şöyle yapıcaz:
```

```python
df.fillna(df.mean()["V1":"V2"]) #v1den v2 ye kadar bu işlemi yap demek v3 ü ellemiyoruz
```

```python
df["V3"].fillna(df["V3"].median()) # bu sefer v3 ü başta belirtiyoruz tek o kaldı zaten
```

```python
#Tüm değişkenler için üçüncü yol
df.where(pd.notna(df), df.mean(), axis = "columns") # ne yapıyor? yakalamış olduğu df içerisindeki eksik değerlerin yerine,
# o değişkenlerin ortalamasını yerleştirmiş oluyor.
```
